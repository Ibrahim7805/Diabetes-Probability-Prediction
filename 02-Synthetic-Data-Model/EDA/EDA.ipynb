```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('new_diabetes_dataset.csv')
data.head()
```

```python
data.info()
```

# Distribution of diagnosed

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='diagnosed_diabetes', hue='diagnosed_diabetes', palette='viridis')
plt.title('Distribution of Diagnosed Diabetes',
          fontsize=20, fontweight='bold', pad=20)
plt.xlabel('Diagnosed Diabetes')
plt.ylabel('Count')
plt.legend(title='Diabetes', labels=['Not Diagnosed (0)', 'Diagnosed (1)'])
plt.show()
```

# Distribution of diagnosed diabetes by age

```python
plt.figure(figsize=(12, 7))

sns.kdeplot(data=data[data['diagnosed_diabetes'] == 0], x='age', fill=True, label='Not Diagnosed (0)', color='teal', alpha=0.5)
sns.kdeplot(data=data[data['diagnosed_diabetes'] == 1], x='age', fill=True, label='Diagnosed (1)', color='orange', alpha=0.5)

plt.title('Age Density Distribution by Diabetes Status', fontsize=22, fontweight='bold', pad=20)
plt.xlabel('Age', fontsize=16)
plt.ylabel('Density', fontsize=16)
plt.legend(fontsize=12)

plt.show()
```
